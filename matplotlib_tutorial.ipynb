## A simple example

## Either oo interface or pyplot interface is ok

### OO interface

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

x = np.linspace(0, 2)
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, np.power(x, 2), label='quadratic')
ax.plot(x, np.power(x, 3), label='cubic')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('simple')
ax.legend()
% reset -f
```

### Pyplot interface

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

x = np.linspace(0, 2)
plt.plot(x, x, label='linear')
plt.plot(x, np.power(x, 2), label='quadratic')
plt.plot(x, np.power(x, 3), label='cubic')
plt.xlabel('x')
plt.ylabel('y')
plt.title('simple')
plt.legend()

% reset -f
```

## Use subplots

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

fig, axs = plt.subplots(1, 2)
x = np.linspace(0, 2)
axs[0].plot(x, np.power(x, 2), label='quadratic')
axs[1].plot(x, np.power(x, 3), label='cubic')

% reset -f
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

fig, axs = plt.subplots(2, 2)
x = np.linspace(0, 2)
axs[0, 0].plot(x, np.power(x, 2), label='quadratic')
axs[1, 1].plot(x, np.power(x, 3), label='cubic')
axs[0, 0].legend()
axs[1, 1].legend()
% reset -f
```

### Share axis

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

fig, axs = plt.subplots(1, 2, sharey=True)
x = np.linspace(0, 2)
axs[0].plot(x, np.power(x, 2), label='quadratic')
axs[1].plot(x, np.power(x, 3), label='cubic')
axs[0].legend()
axs[1].legend()

% reset -f
```

## Plotting with keyword strings

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)  ## c defines the color of dots, s defines the size of dots
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()

% reset -f
```

## Plotting with categorical variables

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()

% reset -f
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

fig, axs = plt.subplots(1, 3)
fig.set_figheight(3)
fig.set_figwidth(9)
fig.suptitle('Categorical Plotting')

axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

% reset -f
```

## Controlling line properties

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd


def gaussian(x, mu, sigma):
    return 1 / (np.sqrt(2 * np.pi)) / sigma * (np.exp(-np.power(x - mu, 2) / np.power(sigma, 2)))


x = np.linspace(-3, 3)
fig, ax = plt.subplots()
line = ax.plot(x, gaussian(x, 0, 1), marker='+', mec='r')
plt.setp(line)
% reset -f
```

## Working with multiple figures and axes

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd


def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()  # you can ignore and plt will create a figure automatically
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')
plt.show()  # you can ignore in jupyter

% reset -f
```

## Working with text

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd


def gaussian(x, mu, sigma):
    return 1 / (np.sqrt(2 * np.pi)) / sigma * (np.exp(-np.power(x - mu, 2) / np.power(sigma, 2)))


x = np.random.normal(size=500)
fig, ax = plt.subplots()
ax.hist(x, 50, density=True, color='b')
ax.set_xlabel('x')
ax.set_ylabel('p')
y = np.linspace(-3, 3)
ax.plot(y, gaussian(y, 0, 1), c='r')

ax.set_title('Gaussian')
ax.text(1, 0.4, r'$\mu=0,\sigma=1$')
% reset -f
```

## Annotation

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

x = np.linspace(-2 * np.pi, 2 * np.pi)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.annotate('local max', xy=(np.pi / 2, 1), xytext=(3, 1), arrowprops=dict(facecolor='black', shrink=0))

% reset -f
```

## Surface plots

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import pandas as pd

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)
print(X)
print(np.shape(X))
print(R)
print(Z)

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
surf = ax.plot_surface(X, Y, Z)


% reset -f
```
